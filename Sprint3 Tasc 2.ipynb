```python
# Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)


df['late_arrival'] = np.where(df['CRSArrTime'] > df['ArrTime'], 'Yes', 'No')

df['delay'] = ((df['DepDelay'])+(df['ArrDelay'])) #total delay column for a flight (Departure + Arrival delay)

df['Date']= pd.to_datetime(pd.DataFrame({'year':df['Year'],
                                         'month':df['Month'],'day':df['DayofMonth'], 
                                         'hour':df['DepTime']})) #date column
```

```python
# Agrupar los datos por aerolínea y calcular el retraso acumulado
grouped = df.groupby("UniqueCarrier").agg({"DepDelay": "sum"})

# Ordenar los datos por retraso acumulado y tomar la primera fila (aerolínea con mayor retraso acumulado)
result = grouped.sort_values("DepDelay", ascending=False).head(10)

# Mostrar los resultados
print(result)
```

```python
# Agrupamos los datos por la columna 'UniqueCarrier' y calculamos la suma de la columna 'Distance'
grouped_distance = df.groupby('UniqueCarrier')['Distance'].sum()

# Ordenamos los datos en orden descendente y obtenemos la primera fila para encontrar la aerolínea con la mayor distancia total
most_distance_carrier = grouped_distance.sort_values(ascending=False).index[0]

# Obtenemos la distancia total de la aerolínea con la mayor distancia total
most_distance = grouped_distance.loc[most_distance_carrier]

# Agrupamos los datos por la columna 'UniqueCarrier' y calculamos la suma de la columna 'DepDelay'
grouped_delay = df.groupby('UniqueCarrier')['DepDelay'].sum()

# Ordenamos los datos en orden descendente y obtenemos la primera fila para encontrar la aerolínea con el mayor retraso total
most_delay_carrier = grouped_delay.sort_values(ascending=False).index[0]

# Obtenemos el retraso total de la aerolínea con el mayor retraso total
most_delay = grouped_delay.loc[most_delay_carrier]

# Imprimimos los resultados
print("La aerolinea con la mayor distancia total es {} con una distancia de {} ".format(most_distance_carrier, most_distance))
print("La aerolinea con el mayor retraso total es {} con un retraso de {}".format(most_delay_carrier, most_delay))
```
